```python
import pandas as pd
import numpy as np
import matplotlib

matplotlib.rcParams['font.sans-serif'] = ['Songti SC']

df = pd.read_csv('./data/hero.csv')
```

```python
# 1. 英雄统计，绘制柱状图
hero_type_df_groupby = df.groupby(["hero_type"])
hero_type_df_ser = hero_type_df_groupby.size()
hero_type_df_ser.plot(kind='bar')
```

```python
def get_max_in_group(param):
    max_life = hero_type_df_groupby[["hero_type", "hero_name", param]].apply(lambda hero: hero[hero[param] == hero[param].max()])
    return max_life[max_life['hero_type'] == '坦克']
```

```python
# 2. 分组操作，坦克下面基础生命最大的英雄
get_max_in_group("max_life")
```
